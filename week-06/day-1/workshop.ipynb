```python
sorted_rating = df.sort_values(by = ['Rating'], ascending = False)
```

```python
first_q = sorted_rating[sorted_rating['Rating'] >= sorted_rating['Rating'].quantile(q = 0.25)]
```

```python
first_q
```

```python
def convert_size(value):
    if value == 'Varies with device':
        return 0.99999
    elif 'k' in value:
        new_size = float(value[:-1])
        return new_size
    elif 'M' in value:
        new_size = float(value[:-1]) * 1024
        return new_size
df['Ture Size'] = df['Size'].apply(convert_size)
def review_ppl(value):
    new_reviews = float(value)
    return new_reviews
df['Reviews post'] = df['Reviews'].apply(review_ppl)
size_df = df[df['Size'] != 'Varies with device']
```

```python
sorted_size = size_df.sort_values(by = ['Ture Size'], ascending = False)
sorted_size
```

```python
large_size = sorted_size[sorted_size['Ture Size'] >= (50 * 1024)]
large_size
```
